### Observations

| Observations | \( X_1 \) | \( X_2 \) | \( X_3 \) | \( X_4 \) | \( Y \) (cible \( t_i \)) |
|--------------|-----------|-----------|-----------|-----------|--------------------------|
| \( P_1 \)    | 1         | 1         | 1         | 1         | 1                        |
| \( P_2 \)    | 0         | 1         | 0         | 0         | 0                        |
| \( P_3 \)    | 0         | 0         | 1         | 0         | 1                        |
| \( P_4 \)    | 0         | 0         | 0         | 1         | 0                        |
| \( P_5 \)    | 0         | 0         | 0         | 0         | 0                        |



### Paramètres initiaux

-  W = [1, 1, 1, 1]  : vecteur des poids
-  b = 0  :biais
- alpha = 1   :taux d'apprentissage


## **2. Itérations d'apprentissage**

### **Étape 1 : Initialisation**
- **Poids initiaux** : W = [1, 1, 1, 1]
- **Biais initial** : b = 0

---

### **Étape 2 : Apprentissage itératif**
Pour chaque observation, si l'erreur n'est pas nulle \\( e \neq 0 \\)
, les poids et le biais sont mis à jour, et le processus reprend depuis la première observation.

#### **Itération 1 :**
 Observation P_1:
1. **Calcul de la sortie :**
   $
   Y_  \text{sortie}=f_\text{hardlim}(W \cdot P_1^T + b) = f_\text{hardlim}(4) = 1.
   $
2. **Erreur :**
   $
   e = t_1 - Y_{\text{sortie}} = 1 - 1 = 0.$


Observation P_2:
1. **Calcul de la sortie :**
  $ Y_{\text{sortie}} = f_\text{hardlim}(W \cdot P_2^T + b) = f_\text{hardlim}(1) = 1.$
2. **Erreur :**
$e = t_2 - Y_{\text{sortie}} = 0 - 1 = -1.$

#**Problème 2: Problème de classification par un ou plusieurs "Simple Linear Perceptron"**



## Problème

Les points rouges et bleus dans le graphique doivent être classifiés à l'aide d'un ou plusieurs perceptrons simples avec une fonction d'activation linéaire, sans utiliser d'implémentations informatiques.

---
#Méthode1 : **Classification avec deux Perceptrons (Méthode Droites Linéaires)**

## **Approche**
Deux perceptrons sont utilisés pour résoudre le problème de classification des points rouges et bleus :
- Chaque perceptron définit une droite de séparation.
- Les sorties des deux perceptrons sont combinées logiquement pour déterminer la classe finale.



### **Perceptron 1 : Droite $x_2 = x_1$**
- **Équation** : $x_2 - x_1 = 0$.
- **Poids et biais** :
  - Poids : $w_1 = 1$, $w_2 = -1$.
  - Biais : $b = 0$.
- **Fonction de sortie** :
$
y_1 =
\begin{cases}
1, & \text{si } x_2 \geq x_1 \\
0, & \text{sinon.}
\end{cases}$



### **Perceptron 2 : Droite $x_2 = -x_1$**
- **Équation** : $x_2 + x_1 = 0$.
- **Poids et biais** :
  - Poids : $w_1 = 1$, $w_2 = 1$.
  - Biais : $b = 0$.
- **Fonction de sortie** :
$
y_2 =
\begin{cases}
1, & \text{si } x_2 \geq -x_1 \\
0, & \text{sinon.}
\end{cases}
$



### **Décision Finale**
La classe finale est déterminée comme suit :
$
\text{Classe finale} =
\begin{cases}
1 \, (\text{Rouge}), & \text{si } y_1 = 1 \text{ et } y_2 = 1 \\
0 \, (\text{Bleu}), & \text{sinon.}
\end{cases}$



## **Vérification**
1. **Point Bleu $(-1, 1)$** :
   - $x_2 - x_1 = 2 \implies y_1 = 1$,
   - $-x_2 + x_1 = -2 \implies y_2 = 0$,
   - Résultat : $y_1 = 1 \text{ et } y_2 = 1 \implies \text{Classe = Bleu.}$

2. **Point Bleu $(1, -1)$** :
   - $x_2 - x_1 = 2 \implies y_1 = 1$,
   - $-x_2 + x_1 = -2 \implies y_2 = 0$,
   - Résultat : $y_1 = 0 \text{ et } y_2 = 1 \implies \text{Classe = Bleu.}$

3. **Point Rouge $(1, 1)$** :
   - $x_2 - x_1 = 0 \implies y_1 = 1$,
   - $-x_2 + x_1 = 0 \implies y_2 = 1$,
   - Résultat : $y_1 = 1 \text{ et } y_2 = 1 \implies \text{Classe = Rouge.}$

4. **Point Rouge $(-1, -1)$** :
   - $x_2 - x_1 = 0 \implies y_1 = 1$,
   - $-x_2 + x_1 = 0 \implies y_2 = 1$,
   - Résultat : $y_1 = 1 \text{ et } y_2 = 0 \implies \text{Classe = Rouge.}$

## Méthode 2 : **Vote majoritaire**

### Approche
- Deux perceptrons sont utilisés pour voter sur la classe d'un point.
- Chaque perceptron calcule une séparation pour une dimension donnée, et la classe finale est décidée par un vote.


1. Deux perceptrons sont définis avec les lignes suivantes :
   - Perceptron 1 : $x_1 = 0$ (sépare les points en fonction de l'axe $x_1$).
   - Perceptron 2 : $x_2 = 0$ (sépare les points en fonction de l'axe $x_2$).

2. Sorties des perceptrons :
   - Perceptron 1 :
$
\text{Sortie}_1 =
\begin{cases}
1 & \text{si } x_1 \geq 0 \\
0 & \text{sinon.}
\end{cases}
$
   - Perceptron 2 :
$
\text{Sortie}_2 =
\begin{cases}
1 & \text{si } x_2 \geq 0 \\
0 & \text{sinon.}
\end{cases}
\$

3. Décision finale :
   - La classe est définie par un vote majoritaire :
$
\text{Classe finale} =
\begin{cases}
1 & \text{si } \text{Sortie}_1 + \text{Sortie}_2 = 1 \\
0 & \text{sinon.}
\end{cases}
$

### Vérification
- Point Bleu $ (1, -1) $ :
  - $ \text{Sortie}_1 = 1 $, $ \text{Sortie}_2 = 0 $, $\implies \text{Classe = Bleu.}$
- Point rouge $(1, 1)$  :
  - $\text{Sortie}_1 = 1 $, $ \text{Sortie}_2 = 1 $, $\implies \text{Classe = Rouge.}$
- Point Bleu $ (-1, 1) $ :
  - $ \text{Sortie}_1 = 0 $, $ \text{Sortie}_2 = 1 $, $\implies \text{Classe = Bleu.}$
- Point rouge $(-1, -1)$  :
  - $\text{Sortie}_1 = 1 $, $ \text{Sortie}_2 = 1 $, $\implies \text{Classe = Rouge.}$

## Méthode 3 : Produit $z = x_1 \times x_2 $

### Approche



1. Deux neurones sont définis :
   - **Neurone 1 :** Prend directement $ x_1 $ comme entrée.
   - **Neurone 2 :** Calcule le produit $ z = x_1 \times x_2 $

   avec les poids et biais définis comme suit :
- Poids : $ w_1 = 0 $, $ w_2 = 1 $,
- Biais : $ b = 0 $.
2. Sortie du perceptron :


   - La classe est déterminée selon la règle suivante :
$
\text{Classe finale} =
\begin{cases}
1 \, (\text{Rouge}), & \text{si } z = x_1 \times x_2 \geq 0, \\
0 \, (\text{Bleu}), & \text{sinon.}
\end{cases}
$



### Vérification
- **Point Bleu $(1, -1) $** :
  - $ z = 1 \times -1 = -1 $,
  - $ z < 0 $ , $\implies \text{Classe = Bleu.}$.

- **Point Rouge $ (1, 1) $** :
  - $z = 1 \times 1 = 1 $,
  - $ z \geq 0 $,  $\implies \text{Classe = Rouge.}$ .

- **Point Bleu $ (-1, 1) $** :
  - $ z = -1 \times 1 = -1 $,
  - $ z < 0 $, $\implies \text{Classe = Bleu.}$ .

- **Point Rouge $ (-1, -1) $** :
  - $ z = -1 \times -1 = 1 $,
  - $ z \geq 0 $, $\implies \text{Classe = Rouge.}$.
